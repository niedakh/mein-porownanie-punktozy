```python
import pandas as pd
```

```python
old_p = pd.read_excel('./20210218.xlsx', sheet_name=0)
old_c = pd.read_excel('./20210218.xlsx', sheet_name=1)
new_p = pd.read_excel('./20211201.xlsx', sheet_name=0)
new_c = pd.read_excel('./20211201.xlsx', sheet_name=1)
```

```python
old_inf_p = old_p[old_p['informatyka techniczna i telekomunikacja'].notnull()].iloc[:,1:9]
new_inf_p = new_p[new_p['informatyka techniczna i telekomunikacja'].notnull()].iloc[:,1:9]
```

```python
old_inf_p.columns = ['UID', 'Tytuł 1', 'ISSN', 'E-ISSN', 'Tytuł 2', 'ISSN1', 'EISSN1', 'Punkty']
new_inf_p.columns = ['UID', 'Tytuł 1', 'ISSN', 'E-ISSN', 'Tytuł 2', 'ISSN1', 'EISSN1', 'Punkty']
```

```python
def get_title(row):
    if str(row['Tytuł 1']) != 'nan':
        return row['Tytuł 1']
    return row['Tytuł 2']
```

```python
old_inf_p['Tytuł']=old_inf_p.apply(get_title, axis=1)
new_inf_p['Tytuł']=new_inf_p.apply(get_title, axis=1)
```

```python
main_columns = ['UID', 'Tytuł', 'Punkty']
```

```python
merged_p=old_inf_p[main_columns].merge(new_inf_p[main_columns], on='UID', how='right')
```

```python
changes = merged_p[merged_p['Punkty_x']!=merged_p['Punkty_y']][['Tytuł_y', 'Punkty_x', 'Punkty_y']].copy().sort_values('Punkty_x', ascending=False)
```
